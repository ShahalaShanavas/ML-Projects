```python
import numpy as np
import pandas as pd
df=pd.read_csv('/content/Salary_Data.csv')
df
```

```python
x=df.iloc[:,:-1] #.values to set as array is not needed in regression
x
```

```python
y=df.iloc[:,-1]
y
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y)
plt.xlabel('Experience',color='r')
plt.ylabel('Salary',color='r')
plt.title('Experience-Salary',color='g',fontsize='20')

```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
y_test
```

```python
#model creation
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred

```

```python
plt.scatter(x_train,y_train,color='b') #to plot training points
plt.plot(x_test,y_pred,color='r') #to plot regression line
```

```python
print(model.predict([[7.8]]))
```

```python
data={'Actual':y_test,'Predicted':y_pred,'Difference':y_test-y_pred} #dictionary
df1=pd.DataFrame(data)
df1
```

```python
#linear regression works based on y=mx+c
#print slope and y intercept on x
print("Slope is ",model.coef_)
```

```python
print("Constant is ",model.intercept_)
```

```python
#performance evaluation
#mean_absolute_error
from sklearn.metrics import mean_absolute_error
print("Error is: ",mean_absolute_error(y_test,y_pred))
```

```python
#mean_absolute_percentage_error
from sklearn.metrics import mean_absolute_percentage_error
print("Error is: ",mean_absolute_percentage_error(y_test,y_pred))

#only 6% error -  good model
#can go upto 30% for good model
```

```python
#mean squared error
from sklearn.metrics import mean_squared_error
print("Error is: ",mean_squared_error(y_test,y_pred))
```

```python
#root mean squared error
#it does not have a function
SE=mean_squared_error(y_test,y_pred)
print("Error is: ",np.sqrt(SE))
```

```python
#r2_score: cefficience of determination #it should be high for a good model 70 above
from sklearn.metrics import r2_score
print("Error is: ",r2_score(y_test,y_pred))
```
