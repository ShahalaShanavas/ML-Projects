# **Training Data**

```python
#plotting count v/s unique values of object columns : Location,Fuel_Type,Transmission,Owner_Type
x=df['Location'].unique()
y=list(df['Location'].value_counts())
plt.bar(x,y,color='g')

plt.xlabel('Location',color='r',family='serif',fontsize=15)
plt.ylabel('Count',color='r',family='serif',fontsize=15)
plt.xticks(rotation=90)
plt.title('Location-Count',fontsize='18',color='k',family='serif',weight='bold')
```

```python
x=df['Fuel_Type'].unique()
y=list(df['Fuel_Type'].value_counts())
plt.bar(x,y,color='y')

plt.xlabel('Fuel Type',color='b',family='serif',fontsize=15)
plt.ylabel('Count',color='b',family='serif',fontsize=15)
plt.title('Fuel Type-Count',fontsize='18',color='k',family='serif',weight='bold')
```

```python
x=df['Transmission'].unique()
y=list(df['Transmission'].value_counts())
plt.pie(y,labels=x,autopct='%1.1f%%',colors=['orange','red'])
plt.title('Transmission Type',fontsize='18',color='k',family='serif',weight='bold')
plt.legend(bbox_to_anchor=(1.3,0.14))
```

```python
x=df['Owner_Type'].unique()
y=list(df['Owner_Type'].value_counts())
plt.pie(y,labels=x,colors=['orange','blue','green','red'])
plt.legend(bbox_to_anchor=(1,1))
plt.title('Owner Type',family='sans-serif',weight='bold',fontsize='20')
```

```python
df['Name'].value_counts()
```

```python
#adding a new column 'Brand' from the column 'Name'
df['Brand']=df['Name'].apply(lambda x:x.split()[0])
```

```python
df['Brand'].value_counts()
```

```python
#Encoding : get_dummies
df1=pd.get_dummies(df[['Location','Fuel_Type','Transmission','Owner_Type','Brand']],dtype=int,drop_first=True)
df1
```

```python
#Joining 2 Dataframes

dfe=pd.concat([df,df1],axis=1)
dfe
```

```python
#Removing unwanted columns

dfe.drop(['Unnamed: 0', 'Name','Location','Fuel_Type','Transmission', 'Owner_Type','New_Price','Brand'],axis=1,inplace=True)
dfe
```

```python
dfe.dtypes
```

```python
#replaceing units
dfe['Mileage']=dfe['Mileage'].str.replace('km/kg','')
dfe['Mileage']=dfe['Mileage'].str.replace('kmpl','')
dfe['Engine']=dfe['Engine'].str.replace('CC','')
dfe['Power']=dfe['Power'].str.replace('bhp','')
```

```python
#replacing 'null' with '0'
dfe['Mileage']=dfe['Mileage'].str.replace('null','0')
dfe['Power']=dfe['Power'].str.replace('null','0')
dfe['Engine']=dfe['Engine'].str.replace('null','0')
dfe.dtypes
```

```python
#Datatype conversion
dfe['Mileage']=dfe['Mileage'].astype(float)
dfe['Engine']=dfe['Engine'].astype(float)
dfe['Power']=dfe['Power'].astype(float)
dfe.dtypes
```

```python
#The mileage, engine, and power of a car cannot be 0, so it has to be considered a missing value and has to be handled.
#replacing 0 with NaN
dfe.loc[dfe.Engine==0,'Engine']=np.NaN
dfe.loc[dfe.Mileage==0,'Mileage']=np.NaN
dfe.loc[dfe.Power==0,'Power']=np.NaN
```

```python
#missing values
dfe.isna().sum()
```

```python
#handling missing values
dfe.fillna(dfe['Mileage'].mean(),inplace=True)
dfe.fillna(dfe['Engine'].mean(),inplace=True)
dfe.fillna(dfe['Power'].mean(),inplace=True)
dfe.fillna(dfe['Seats'].mode()[0],inplace=True)
dfe.isna().sum()
```

```python
#seperate input and output
X=dfe.drop(['Price'],axis=1)
X.ndim
```

```python
Y=dfe['Price']
Y
```
