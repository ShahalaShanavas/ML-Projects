```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_csv('/content/heightweight.csv')
df
```

```python
x=df.iloc[:,:-1]
x
```

```python
y=df.iloc[:,-1]
y
```

```python
plt.scatter(x,y)
plt.xlabel('Height(in m)',color='r')
plt.ylabel('Weight(in kg)',color='r')
plt.title('Height-Weight Graph',color='y',fontsize='20')
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
y_test
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred
```

```python
#to plot training points
plt.scatter(x_train,y_train,color='g')
#to plot regression line
plt.plot(x_test,y_pred,color='b')

```

```python
print(model.predict([[1.67]]))
```

```python
data={'actual':y_test,'predicted':y_pred,'difference':y_test-y_pred}
df1=pd.DataFrame(data)
df1
```

```python
#linear regression works based on y=mx+c
#print slope and y intercept on x
print("Slope is: ",model.coef_)
```

```python
print("constant is: ",model.intercept_)
```

```python
#performance evaluation

from sklearn.metrics import mean_absolute_error
print("MAE is: ", mean_absolute_error(y_test,y_pred))
```

```python
#mean_absolute_percentage_error
from sklearn.metrics import mean_absolute_percentage_error
print("MAPE is: ",mean_absolute_percentage_error(y_test,y_pred))
```

```python
#mean squared error
from sklearn.metrics import mean_squared_error
print("MSE is: ",mean_squared_error(y_test,y_pred))
```

```python
#root mean squared error
data=mean_squared_error(y_test,y_pred)
print("RMSE is: ",np.sqrt(data))
```

```python
#r2_score
from sklearn.metrics import r2_score
print("r2_score is: ",r2_score(y_test,y_pred))
```
